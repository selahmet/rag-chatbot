### Basit Bir ML Pipeline Ã–rneÄŸi

```python
# Cell 1: generate synthetic credit risk data (more separable, ~18% default)
import warnings 
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd

rng = np.random.default_rng(42)
n = 6000

# Core drivers
income = rng.lognormal(mean=9.5, sigma=0.5, size=n)            # annual income (TRY)
age = rng.integers(20, 70, size=n)                             # age
debt_ratio = np.clip(rng.normal(0.35, 0.15, size=n), 0, 1)     # debt/income
late_payments = rng.poisson(lam=1.3, size=n)                   # # of late pays in last 12m
num_accounts = rng.integers(1, 12, size=n)                     # # of credit accounts
util_rate = np.clip(rng.normal(0.5, 0.2, size=n), 0, 1)        # revolving utilization

# A more separable logit: stronger weights on risk drivers + mild interactions
logit = (
    -2.3
    - 0.00005 * income
    + 0.015 * (50 - age)
    + 3.1 * debt_ratio
    + 0.55 * late_payments
    + 1.3 * util_rate
    + 0.35 * (num_accounts > 8)
    + 0.8 * (debt_ratio * util_rate)            # interaction increases risk
)

p = 1 / (1 + np.exp(-logit))
default = rng.binomial(1, p)

df = pd.DataFrame({
    "income": income,
    "age": age,
    "debt_ratio": debt_ratio,
    "late_payments": late_payments,
    "num_accounts": num_accounts,
    "util_rate": util_rate,
    "default": default.astype(int)
})

df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(context="notebook", style="whitegrid")

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# (A) Target distribution
# Rationale: Always inspect class balance before modeling.
df["default"].value_counts(normalize=True).sort_index().plot(
    kind="bar", ax=axes[0]
)
axes[0].set_xticklabels(["No Default (0)", "Default (1)"], rotation=0)
axes[0].set_title("Target Distribution")
axes[0].set_ylabel("Proportion")

# (B) Binned default rate vs debt_ratio
# Rationale: Directly shows how risk changes across feature ranges.
bins_dr = pd.cut(df["debt_ratio"], bins=np.linspace(0, 1, 11))
rate_dr = df.groupby(bins_dr)["default"].mean()
rate_dr.index = rate_dr.index.astype(str)
axes[1].plot(range(len(rate_dr)), rate_dr.values, marker="o")
axes[1].set_xticks(range(len(rate_dr)))
axes[1].set_xticklabels(rate_dr.index, rotation=45, ha="right")
axes[1].set_title("Default Rate by Debt Ratio (binned)")
axes[1].set_ylabel("Default Rate")
axes[1].set_xlabel("Debt Ratio Bin")

# (C) Correlation heatmap (features only)
# Rationale: Quick scan for multicollinearity / relationships.
corr = df.drop(columns=["default"]).corr(numeric_only=True)
sns.heatmap(corr, cmap="vlag", annot=False, ax=axes[2])
axes[2].set_title("Feature Correlations")

plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, average_precision_score, classification_report, precision_recall_curve
import numpy as np

y = df["default"].astype(int)
X = df.drop(columns=["default"])

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, stratify=y, random_state=42
)

num_features = X_train.columns.tolist()
preprocess = ColumnTransformer(
    transformers=[("num", StandardScaler(), num_features)],
    remainder="drop"
)

clf = Pipeline(steps=[
    ("prep", preprocess),
    ("model", LogisticRegression(
        max_iter=1000,
        class_weight="balanced",   # helps with class imbalance
        random_state=42
    ))
])

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_auc = cross_val_score(clf, X_train, y_train, scoring="roc_auc", cv=cv).mean()

clf.fit(X_train, y_train)
proba = clf.predict_proba(X_test)[:, 1]
roc_auc = roc_auc_score(y_test, proba)
pr_auc = average_precision_score(y_test, proba)


precision, recall, thresh = precision_recall_curve(y_test, proba)
f1_proxy = 2 * (precision * recall) / (precision + recall + 1e-12)
best_idx = np.nanargmax(f1_proxy)
best_threshold = thresh[min(best_idx, len(thresh)-1)]
pred = (proba >= best_threshold).astype(int)

print(f"CV ROC AUC (train): {cv_auc:.3f}")
print(f"Test ROC AUC      : {roc_auc:.3f}")
print(f"Test PR AUC       : {pr_auc:.3f}")
print(f"Chosen threshold  : {best_threshold:.3f}\n")
print("Classification report (with tuned threshold):\n")
print(classification_report(y_test, pred, digits=3))
```

### Basit Bir DL Pipeline Ã–rneÄŸi
**Dataset:** [https://www.kaggle.com/datasets/zalando-research/fashionmnist/data](https://www.kaggle.com/datasets/zalando-research/fashionmnist/data)

```python
# Cell 1: load Fashion-MNIST (from Kaggle), subset, split, normalize
import os, gzip, struct, numpy as np, pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
import tensorflow as tf
import kagglehub

np.random.seed(42); tf.random.set_seed(42)

path = kagglehub.dataset_download("zalando-research/fashionmnist")
root = Path(path)

def read_idx_images(p):
    with gzip.open(p, "rb") as f:
        _, n, r, c = struct.unpack(">IIII", f.read(16))
        return np.frombuffer(f.read(), dtype=np.uint8).reshape(n, r, c)

def read_idx_labels(p):
    with gzip.open(p, "rb") as f:
        _, n = struct.unpack(">II", f.read(8))
        return np.frombuffer(f.read(), dtype=np.uint8)

# detect format
has_idx = any(root.rglob("*-idx*-ubyte.gz"))
has_csv = any(root.rglob("*.csv"))

if has_idx:
    tr_x = next(root.rglob("train-images-idx3-ubyte.gz"))
    tr_y = next(root.rglob("train-labels-idx1-ubyte.gz"))
    te_x = next(root.rglob("t10k-images-idx3-ubyte.gz"))
    te_y = next(root.rglob("t10k-labels-idx1-ubyte.gz"))
    X_train_full = read_idx_images(tr_x)
    y_train_full = read_idx_labels(tr_y)
    X_test_full  = read_idx_images(te_x)
    y_test_full  = read_idx_labels(te_y)
elif has_csv:
    tr = next(p for p in root.rglob("*.csv") if "train" in p.name.lower())
    te = next(p for p in root.rglob("*.csv") if "test"  in p.name.lower())
    df_tr, df_te = pd.read_csv(tr), pd.read_csv(te)
    y_train_full = df_tr["label"].to_numpy(np.uint8)
    X_train_full = df_tr.drop(columns=["label"]).to_numpy(np.uint8).reshape(-1, 28, 28)
    y_test_full  = df_te["label"].to_numpy(np.uint8)
    X_test_full  = df_te.drop(columns=["label"]).to_numpy(np.uint8).reshape(-1, 28, 28)
else:
    raise FileNotFoundError(f"No IDX/CSV files found under {root}")

# use subsets to keep it light
n_train_subset, n_test_subset = 20_000, 5_000
idx = np.random.choice(len(X_train_full), size=n_train_subset, replace=False)
X_small, y_small = X_train_full[idx], y_train_full[idx]
X_test = X_test_full[:n_test_subset]
y_test = y_test_full[:n_test_subset]

# split + normalize + channel dim
X_train, X_val, y_train, y_val = train_test_split(
    X_small, y_small, test_size=0.1, stratify=y_small, random_state=42
)

def to_tensor(x):
    x = x.astype("float32") / 255.0
    return np.expand_dims(x, -1)

X_train, X_val, X_test = map(to_tensor, [X_train, X_val, X_test])
num_classes = int(len(np.unique(y_train)))

print("Shapes:",
      "\n  train:", X_train.shape, y_train.shape,
      "\n  val  :", X_val.shape,   y_val.shape,
      "\n  test :", X_test.shape,  y_test.shape,
      "\n  num_classes:", num_classes)
```

```python
# Cell 2: define & train model
from tensorflow.keras import layers, models, callbacks

def make_cnn(input_shape=(28, 28, 1), num_classes=10):
    m = models.Sequential([
        layers.Conv2D(32, 3, activation="relu", input_shape=input_shape),
        layers.Conv2D(32, 3, activation="relu"),
        layers.MaxPooling2D(), layers.Dropout(0.25),
        layers.Conv2D(64, 3, activation="relu"),
        layers.MaxPooling2D(), layers.Dropout(0.25),
        layers.Flatten(),
        layers.Dense(128, activation="relu"), layers.Dropout(0.3),
        layers.Dense(num_classes, activation="softmax"),
    ])
    m.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])
    return m

model = make_cnn(input_shape=(28,28,1), num_classes=num_classes)
es = callbacks.EarlyStopping(monitor="val_accuracy", patience=3, restore_best_weights=True)

hist = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=10,
    batch_size=128,
    callbacks=[es],
    verbose=1
)
```

```python
# Cell 3: evaluate & plot
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report

test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)
print(f"Test accuracy: {test_acc:.3f} | Test loss: {test_loss:.3f}")

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(hist.history["loss"], label="train")
plt.plot(hist.history["val_loss"], label="val"); plt.title("Loss"); plt.legend()
plt.subplot(1,2,2)
plt.plot(hist.history["accuracy"], label="train")
plt.plot(hist.history["val_accuracy"], label="val"); plt.title("Accuracy"); plt.legend()
plt.tight_layout(); plt.show()

y_pred = model.predict(X_test[:2000], verbose=0).argmax(axis=1)
print(classification_report(y_test[:2000], y_pred, digits=3))
```
